```python
import pandas as pd
import seaborn as sns
```

```python
gameSales = pd.read_csv('../Datasets/vgsales.csv', sep=',')
```

```python
gameSales.head()
```

```python
gameSales['Platform'].value_counts(dropna=False).plot(kind='bar', title="total games per platform", xlabel="Platform", ylabel="Amount")
```

```python
gameSales['Genre'].value_counts(dropna=False).plot(kind='bar', title="total games per genre", xlabel="Genre", ylabel="Amount")
```

```python
columns = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
gameSales[columns].sum().plot(kind='bar', title="Game sales per region", ylabel="Sales (Millions)", xlabel="Region")
```

```python
gameSales['Global_Sales'].plot(kind='box', title= 'Boxplot Sales')
```

```python
gameSales[gameSales['Global_Sales'] < 0.49]['Global_Sales'].plot(kind='box', title= "Boxplot Sales No Outliers")
```

```python
gameSalesYearFiltered = gameSales['Year'].dropna().astype(int)
gameSalesYearFiltered.value_counts(dropna=True).sort_index(ascending=False).plot(kind='bar', title="Amount of games per year", xlabel="year", ylabel="Amount")
```

```python
gameSales['Year'].plot(kind='box', title= 'Boxplot Games Per Year')
```

```python
gameSales[gameSales['Year'] > 1992]["Year"].plot(kind='box', title= "Boxplot Games Per Year No Outliers")
```
