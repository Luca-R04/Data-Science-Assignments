```python
import pandas as pd
import seaborn as sns
```

```python
gameSales = pd.read_csv('../Datasets/vgsales.csv', sep=',')
```

```python
gameSales.head()
```

```python
gameSalesCorrelation = gameSales.corr()
gameSalesCorrelation.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

```python
gameSales.plot(kind ='scatter', x = 'NA_Sales', y = 'Global_Sales', title='Strongest positive correlation')
```

```python
gameSales.plot(kind ='scatter', x = 'Year', y = 'JP_Sales', title='Strongest negative correlation')
```

```python
gameSales.plot(kind ='scatter', x = 'Year', y = 'EU_Sales', title='Weakest correlation')
```
