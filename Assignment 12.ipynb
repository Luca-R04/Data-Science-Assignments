```python
import pandas as pd
import seaborn as sns
```

```python
gameSales = pd.read_csv('../Datasets/vgsales.csv', sep=',')
```

```python
gameSales['Year'] = gameSales['Year'].astype('object')
```

```python
gameSales.dtypes
```

```python
gameSales.head(5)
```

```python
gameSales.groupby('Genre').mean()
```

```python
with sns.plotting_context():
    pllt = sns.barplot(y="Global_Sales", x="Genre", data=gameSales)
    pllt.set_xticklabels(pllt.get_xticklabels(), rotation=45, ha="right")
```

```python
gameSales.groupby('Platform').mean()
```

```python
with sns.plotting_context():
    pllt = sns.barplot(y="Global_Sales", x="Platform", data=gameSales)
    pllt.set_xticklabels(pllt.get_xticklabels(), rotation=66, ha="center")
```

```python
gameSales.groupby('Year').mean()
```

```python
with sns.plotting_context():
    pllt = sns.barplot(y="Global_Sales", x="Year", data=gameSales)
    pllt.set_xticklabels(pllt.get_xticklabels(), rotation=90, ha="center")
```
