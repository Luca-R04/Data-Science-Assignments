```python
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins['species'].value_counts().plot(kind='bar', title="Amount per species", xlabel='Species', ylabel='Amount')
```

```python
penguins['island'].value_counts().plot(kind='bar', title='Amount per island', xlabel='Island', ylabel='Amount')
```

```python
penguins['sex'].value_counts().plot(kind='bar', title='Amount per sex', xlabel='Sex', ylabel='Amount')
```
